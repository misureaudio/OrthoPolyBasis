```python
# Import the modules
from legendre.core import LegendreGenerator
from legendre.sympy_integration import generate_sympy_legendre, get_sympy_legendre_basis
from legendre.numpy_integration import evaluate_numpy_legendre, get_numpy_legendre_basis
from legendre.mpmath_integration import evaluate_mpmath_legendre

# Create a generator instance
gen = LegendreGenerator()

# Get coefficients of P_n(x) - in descending order (highest degree first)
print("P_3 coefficients:", gen.get_coefficients(3))  # [2.5, 0.0, -1.5, 0.0]

# Evaluate at a point
print("P_3(0.5) =", gen.evaluate(0.5, 3))  # 0.125

# Generate a basis of polynomials
basis = gen.generate_basis(5)
for i, coeffs in enumerate(basis):
    print(f"P_{i}: {coeffs}")

# SymPy - symbolic computation
poly = generate_sympy_legendre(3)
print("P_3 (SymPy):", poly)  # 5*x**3/2 - 3*x/2
print("d/dx(P_3):", poly.diff('x'))  # 15*x**2/2 - 3/2

# NumPy - numerical evaluation on arrays
import numpy as np
xs = np.linspace(-1, 1, 10)
values = evaluate_numpy_legendre(xs, 4)
print("P_4 values:", values)

# mpmath - arbitrary precision
result = evaluate_mpmath_legendre('0.333333333333333', 5, dps=50)
print("P_5(1/3) with 50 digits:", result)
```
